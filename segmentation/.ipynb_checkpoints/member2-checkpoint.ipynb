```python
import cv2
import numpy as np
import os

input_folder = "road_member3_sharpened"
output_folder = "alligator_fixed"

os.makedirs(output_folder, exist_ok=True)

for i in range(17, 31):

    img_path = os.path.join(input_folder, f"gaussian_contrast_frame_{i}.jpg")
    image = cv2.imread(img_path)

    if image is None:
        print(f"Frame {i} not found")
        continue

    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (15,15))
    blackhat = cv2.morphologyEx(gray, cv2.MORPH_BLACKHAT, kernel)

    norm = cv2.normalize(blackhat, None, 0, 255, cv2.NORM_MINMAX)

    _, thresh = cv2.threshold(norm, 30, 255, cv2.THRESH_BINARY)


    kernel_small = np.ones((3,3), np.uint8)
    clean = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel_small, iterations=1)

    
    connect = cv2.dilate(clean, kernel_small, iterations=2)

    contours, _ = cv2.findContours(connect, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    output = image.copy()

    for cnt in contours:
        area = cv2.contourArea(cnt)

        if area > 200:

            x, y, w, h = cv2.boundingRect(cnt)
            roi = connect[y:y+h, x:x+w]

            density = cv2.countNonZero(roi) / (w * h)

            if 0.15 < density < 0.5:

                cv2.rectangle(output, (x,y), (x+w,y+h), (0,0,255), 2)
                cv2.putText(output, "Alligator Crack", (x, y-5),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 1)

    cv2.imwrite(os.path.join(output_folder, f"blackhat_{i}.jpg"), blackhat)
    cv2.imwrite(os.path.join(output_folder, f"thresh_{i}.jpg"), thresh)
    cv2.imwrite(os.path.join(output_folder, f"connect_{i}.jpg"), connect)
    cv2.imwrite(os.path.join(output_folder, f"final_{i}.jpg"), output)

print("Detection Completed!")
```
