```python
import cv2
import numpy as np
import os

# Input and output folders
# Input and output folders
input_folder = "road_member2_gaussian_contrast"
output_folder = "segmentation_results"

os.makedirs(output_folder, exist_ok=True)

for i in range(1, 16):

    # Load enhanced frame
    img_path = os.path.join(input_folder, f"frame_{i}.jpg")
    image = cv2.imread(img_path)

    if image is None:
        print(f"Frame {i} not found")
        continue

    # Convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Edge detection for crack segmentation
    edges = cv2.Canny(gray, 50, 150)

    # Morphological operation to connect crack segments
    kernel = np.ones((3,3), np.uint8)
    morph = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel)

    # Find contours
    contours, _ = cv2.findContours(morph, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    output = image.copy()

    for cnt in contours:
        area = cv2.contourArea(cnt)

        # Filter small noise
        if area > 50:
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(output, (x,y), (x+w,y+h), (0,255,0), 2)

    # Save segmentation outputs
    cv2.imwrite(os.path.join(output_folder, f"edges_{i}.jpg"), edges)
    cv2.imwrite(os.path.join(output_folder, f"segmented_{i}.jpg"), morph)
    cv2.imwrite(os.path.join(output_folder, f"crack_detected_{i}.jpg"), output)

print("Segmentation completed for all frames.")
```
