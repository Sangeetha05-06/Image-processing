```python
import cv2
import os
import numpy as np

input_folder = "gaussian_contrast_mem2"
output_folder = "segmentation_results"

os.makedirs(output_folder, exist_ok=True)

min_area = 200
max_area = 5000

for file in os.listdir(input_folder):
    path = os.path.join(input_folder, file)
    img = cv2.imread(path)
    if img is None:
        continue

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gray, (5,5), 0)
    _, mask = cv2.threshold(blur, 120, 255, cv2.THRESH_BINARY_INV)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    marked = img.copy()

    mask_valid = np.zeros_like(mask)

    for cnt in contours:
        area = cv2.contourArea(cnt)
        if min_area < area < max_area:
            x, y, w, h = cv2.boundingRect(cnt)
            cv2.rectangle(marked, (x, y), (x + w, y + h), (0, 0, 255), 2)  # draw box
            cv2.putText(marked, "Pothole", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,0,255), 2)
            cv2.rectangle(mask_valid, (x, y), (x + w, y + h), 255, -1)  # fill mask with box

    cv2.imwrite(os.path.join(output_folder, "mask_" + file), mask_valid)
    cv2.imwrite(os.path.join(output_folder, "detected_" + file), marked)

print("Segmentation completed.")

```

```python
import cv2
import matplotlib.pyplot as plt
import os

folder = "segmentation_results"
frame_idx = 10  # Only frame 10 for report

mask_gaussian_file = f"mask_gaussian_contrast_frame_{frame_idx}.jpg"
detected_gaussian_file = f"detected_gaussian_contrast_frame_{frame_idx}.jpg"

mask_gaussian = cv2.imread(os.path.join(folder, mask_gaussian_file), cv2.IMREAD_GRAYSCALE)
detected_gaussian = cv2.imread(os.path.join(folder, detected_gaussian_file))
detected_gaussian = cv2.cvtColor(detected_gaussian, cv2.COLOR_BGR2RGB)


plt.figure(figsize=(16,5))
plt.subplot(1,4,1)
plt.imshow(mask_gaussian, cmap='gray')
plt.title(f"Mask Gaussian Frame {frame_idx}")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(detected_gaussian)
plt.title(f"Detected Gaussian Frame {frame_idx}")
plt.axis("off")


plt.show()
```
